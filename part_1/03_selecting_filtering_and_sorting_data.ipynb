# 📘 Tutorial 3: Selecting, Filtering, and Sorting Data

> This notebook is provided in a clean, non-executed state for readability and reproducibility.

Once a dataset is in a DataFrame, you rarely plot every column and every row. You select variables, filter observations, sort them into a meaningful order, and then extract x and y columns.

---

## 🎯 Learning objectives

By the end of this notebook, you should be able to:

- select one or more columns,
- select rows with `loc` and `iloc`,
- filter rows with boolean conditions,
- sort a DataFrame,
- extract x and y columns for later plotting,
- avoid common filtering mistakes.


## 🔧 Setup


```python
import pandas as pd
```

## 1. Create a small working table


```python
measurements = pd.DataFrame(
    {
        "sample_id": ["S01", "S01", "S01", "S02", "S02", "S02", "S03", "S03", "S03"],
        "condition": ["control", "control", "control", "treated", "treated", "treated", "treated", "treated", "treated"],
        "time_min": [0, 10, 5, 0, 10, 5, 0, 10, 5],
        "replicate": [1, 1, 1, 1, 1, 1, 2, 2, 2],
        "response": [0.05, 0.28, 0.16, 0.06, 0.39, 0.23, 0.04, 0.37, 0.21],
    }
)

measurements
```

The rows are deliberately not sorted by time. That lets us practise sorting before extracting plotting columns.


## 2. Select columns


```python
measurements["response"]
```

```python
measurements[["sample_id", "time_min", "response"]]
```

A single column returns a Series. A list of columns returns a DataFrame.


## 3. Select rows with `iloc`


```python
measurements.iloc[0]
```

```python
measurements.iloc[0:3]
```

`iloc` is position-based. It does not care about row labels; it uses integer positions.


## 4. Select rows and columns with `loc`


```python
measurements.loc[0:3, ["sample_id", "condition", "response"]]
```

`loc` is label-based. With the default integer index, it can look similar to `iloc`, but the meaning is different. The end label is included in this `loc` slice.


## 5. Boolean filtering


```python
treated = measurements["condition"] == "treated"
treated
```

```python
measurements.loc[treated]
```

```python
late_treated = measurements.loc[
    (measurements["condition"] == "treated") & (measurements["time_min"] >= 5)
]

late_treated
```

Use parentheses around each comparison when combining conditions with `&` or `|`.


## 6. Sorting


```python
sorted_measurements = measurements.sort_values(
    ["sample_id", "time_min", "replicate"]
).reset_index(drop=True)

sorted_measurements
```

Sorting makes the row order meaningful. That matters when extracting a time series for plotting.


## 7. Extract x and y columns


```python
sample_s02 = sorted_measurements.loc[sorted_measurements["sample_id"] == "S02"]

x = sample_s02["time_min"].to_numpy()
y = sample_s02["response"].to_numpy()

print("x:", x)
print("y:", y)
```

```python
assert x.shape == y.shape
assert list(x) == sorted(x)
```

The result is not a Matplotlib figure yet. It is the clean input a plotting function would need later.


## 8. Common filtering mistakes

The most common mistakes are:

- using `and` instead of `&` for pandas conditions,
- forgetting parentheses around comparisons,
- filtering and then accidentally modifying a view,
- sorting visually but not saving the sorted result,
- extracting x and y before checking that rows are ordered.


```python
filtered = measurements.loc[measurements["condition"] == "treated"].copy()
filtered["response_percent"] = filtered["response"] * 100

filtered.head()
```

The `.copy()` call is a useful habit when you plan to add or modify columns after filtering.


## 9. Checkpoint: build a plot-ready subset

Task: create a table containing only control rows with the columns `time_min` and `response`, sorted by time.


```python
control_plot_data = (
    measurements.loc[measurements["condition"] == "control", ["time_min", "response"]]
    .sort_values("time_min")
    .reset_index(drop=True)
)

control_plot_data
```

## 🧭 Key takeaways

- Use square brackets to select columns.
- Use `iloc` for position-based selection and `loc` for label-based selection.
- Build boolean filters explicitly and combine them with `&` or `|`.
- Sort before extracting ordered x and y arrays.
- Use `.copy()` when you filter and then plan to modify the result.
