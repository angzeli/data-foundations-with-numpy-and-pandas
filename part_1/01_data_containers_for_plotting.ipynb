# 📘 Tutorial 1: Data Containers for Plotting

This tutorial introduces the three data containers that will appear throughout the series: **NumPy arrays**, **pandas Series**, and **pandas DataFrames**.

Plotting usually starts from a simple question: what should go on the x-axis, what should go on the y-axis, and what labels describe the observations? The answer depends on whether the data is best represented as a plain array or as a labelled table.

## 🔧 Setup


```python
import numpy as np
import pandas as pd
```

## 1. Why data containers matter

A plot is not created from vague data. It is created from concrete values:

- x-values,
- y-values,
- optional groups, labels, or metadata,
- and sometimes uncertainty summaries such as error bars.

NumPy and pandas give us different containers for keeping those values organised.


### A plain NumPy array

A NumPy array is a compact numerical container. It is excellent when the data is already clean, numeric, and position-based.


```python
time_min = np.array([0, 5, 10, 15, 20])
response = np.array([0.05, 0.18, 0.31, 0.42, 0.51])

print(type(time_min))
print(time_min)
print(response)
```

```python
print("time shape:", time_min.shape)
print("response shape:", response.shape)
```

These two arrays already look like the raw material for a line plot: one array for x and one array for y.

The limitation is that the meaning of each array lives mostly in the variable name. The values themselves do not carry column labels, units, or sample identifiers.


### A pandas Series

A Series is a labelled one-dimensional object. It is useful when one column of data needs a name and an index.


```python
response_series = pd.Series(
    [0.05, 0.18, 0.31, 0.42, 0.51],
    name="response",
    index=[0, 5, 10, 15, 20],
)

response_series
```

```python
print("Series name:", response_series.name)
print("Index values:", response_series.index.to_list())
print("Shape:", response_series.shape)
```

The Series now knows two things: the values and the index labels. That can be useful, but most practical datasets contain multiple named columns.


### A pandas DataFrame

A DataFrame is a labelled table. It is usually the safest container for raw or semi-processed tabular data because each column can carry a clear name.


```python
measurements = pd.DataFrame(
    {
        "sample_id": ["S01", "S01", "S01", "S01", "S01"],
        "time_min": [0, 5, 10, 15, 20],
        "response": [0.05, 0.18, 0.31, 0.42, 0.51],
    }
)

measurements
```

```python
print("DataFrame shape:", measurements.shape)
print("Column names:", measurements.columns.to_list())
```

The shape `(5, 3)` means five rows and three columns. For plotting, the important point is that the candidate x and y variables are now explicit columns.


## 2. Shape basics

Shape tells you how data is arranged. Many plotting mistakes begin with arrays that have unexpected dimensions.


```python
one_dimensional = np.array([1, 2, 3, 4])
two_dimensional = np.array([[1, 2], [3, 4], [5, 6]])

print("1D array:", one_dimensional.shape)
print("2D array:", two_dimensional.shape)
```

```python
table = pd.DataFrame(
    {
        "time_min": [0, 5, 10],
        "sample_A": [0.10, 0.24, 0.36],
        "sample_B": [0.08, 0.19, 0.29],
    }
)

print(table.shape)
table
```

In a table, rows usually represent observations and columns represent variables. That convention will become especially important when we reshape data later.


## 3. Extracting x and y values

Even when the data starts in a DataFrame, many plotting functions ultimately receive arrays or array-like columns.


```python
x = measurements["time_min"].to_numpy()
y = measurements["response"].to_numpy()

print("x:", x)
print("y:", y)
```

```python
assert x.shape == y.shape
print("x and y have matching shapes:", x.shape)
```

A small shape check is often worth doing before plotting. If `x` and `y` do not align, the figure will either fail or tell the wrong story.


## 4. When an array is enough

A NumPy array is enough when:

- the values are already numeric,
- the meaning of each axis is obvious,
- there are no important row labels or metadata,
- and you are working with a single measurement series.


```python
calibration_signal = np.array([0.0, 0.21, 0.41, 0.61, 0.82])
concentration_mM = np.array([0, 1, 2, 3, 4])

slope = (calibration_signal[-1] - calibration_signal[0]) / (
    concentration_mM[-1] - concentration_mM[0]
)

print("Approximate slope:", slope)
```

This is compact and clear because the example is small and already tidy.

Once you have multiple samples, replicates, conditions, or units, a DataFrame usually becomes safer.


## 5. When a DataFrame is better

A DataFrame is better when values need context.


```python
replicate_data = pd.DataFrame(
    {
        "sample_id": ["S01", "S01", "S02", "S02"],
        "condition": ["control", "control", "treated", "treated"],
        "replicate": [1, 2, 1, 2],
        "response": [0.42, 0.44, 0.67, 0.71],
    }
)

replicate_data
```

```python
replicate_data.groupby("condition")["response"].mean()
```

The group labels make the calculation readable. We did not have to remember that the first two positions are controls and the last two positions are treated samples.

## 🧭 Key takeaways

- NumPy arrays are compact containers for clean numeric values.
- pandas Series add labels to one-dimensional data.
- pandas DataFrames are the usual starting point for tabular data.
- Plotting later will usually need clean x/y arrays or tidy DataFrames.
- Shape and column names are not cosmetic; they protect the meaning of the data.
