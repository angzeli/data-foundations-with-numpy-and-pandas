Raw data often arrives as CSV files. Before cleaning, filtering, or plotting, the first job is to read the file without damaging it and inspect what pandas actually received.

By the end of this notebook, you should be able to:

- read a CSV file with `pd.read_csv`,

- use relative file paths,

## 🔧 Setup


```python
from pathlib import Path

import pandas as pd
```

## 1. Locate the data file

The notebook should not depend on a machine-specific absolute path. A small project-root helper keeps the path relative while still allowing the notebook to run from either the repository root or the `part_1` folder.


```python
project_root = Path.cwd()
if not (project_root / "data" / "part1").exists():
    project_root = project_root.parent

data_dir = project_root / "data" / "part1"
csv_path = data_dir / "raw_measurements.csv"

print(csv_path)
print("File exists:", csv_path.exists())
```

## 2. Read the CSV

`pd.read_csv` turns the CSV file into a DataFrame. At this stage, avoid changing the raw file by hand. Keep the raw file as evidence and make cleaning steps explicit in code.


```python
raw_data = pd.read_csv(csv_path)
raw_data
```

## 3. Inspect the first and last rows

`head` and `tail` are quick checks. They help answer: does the file look like the file we expected?


```python
raw_data.head()
```

```python
raw_data.tail()
```

## 4. Inspect shape and columns

Shape tells us the number of rows and columns. Column names tell us what pandas thinks each variable is called.


```python
print("Shape:", raw_data.shape)
print("Columns:", raw_data.columns.to_list())
```

This file has one row per measurement and columns for the sample identifier, condition, time, response, and notes.


## 5. Inspect dtypes

Dtypes are one of the fastest ways to spot a problem. A column that looks numeric but has dtype `object` often contains text, missing markers, or mixed formatting.


```python
raw_data.dtypes
```

```python
raw_data.info()
```

The `absorbance` column should be numeric for plotting, but pandas has read it as an object column. That is a sign to inspect it before using it as y-values.


## 6. Identify suspicious columns

We can inspect object columns and look for values that would not convert cleanly to numbers.


```python
object_columns = raw_data.select_dtypes(include="object").columns.to_list()
object_columns
```

```python
raw_data["absorbance"].unique()
```

```python
numeric_probe = pd.to_numeric(raw_data["absorbance"], errors="coerce")
raw_data.loc[numeric_probe.isna(), ["sample_id", "time_min", "absorbance", "operator_note"]]
```

This is only an inspection step. We are not cleaning the column yet. The important habit is to make the issue visible before deciding how to handle it.


## 7. Raw-data discipline

When a raw CSV contains suspicious values, it is tempting to open the file and fix them manually. That is risky because it hides the cleaning decision.

A better workflow is:

- keep the raw file unchanged,
- read it into pandas,
- make cleaning decisions in code,
- save a processed version separately.


## 8. Checkpoint: inspect before cleaning

Before cleaning a new CSV, ask:

- How many rows and columns were read?
- Are the column names what I expected?
- Are numeric-looking columns actually numeric?
- Are there notes, flags, or missing markers?
- Where will the processed output be saved later?


## 🧭 Key takeaways

- Use `pd.read_csv` to read raw CSV files into DataFrames.
- Use `head`, `tail`, `shape`, `columns`, `dtypes`, and `info` before cleaning.
- Numeric-looking columns can still be text columns.
- Keep raw files unchanged and express cleaning decisions in code.
- Relative paths make notebooks easier to share and rerun.
